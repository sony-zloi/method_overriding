```python
class Car:
    """
    К уже реализованному классу «Автомобиль» добавьте
    конструктор, а также необходимые перегруженные методы.
    """

    def __init__(self):
        self.name = input('Name: ')
        self.year = input('Year: ')
        self.vendor = input('Vendor: ')
        self.power = input('Power of engine: ')
        self.color = input('Color: ')
        self.cost = input('Cost: ')

#     def __init__(self, name, year, vendor, power, color, cost):
#         self.name = name
#         self.year = year
#         self.vendor = vendor
#         self.power = power
#         self.color = color
#         self.cost = cost
    
    def __str__(self):
        return f"""name: {self.name.title()} \nyear: {self.year} \nvendor: {self.vendor}
power: {self.power} \ncolor: {self.color} \ncost: {self.cost}\n"""

    def __repr__(self):
        return str(self)

    def getName(self):
        return self.name

    def setName(self, value):
        if not value:
            raise ValueError
        self.name = value

    def getYear(self):
        return self.year

    def setYear(self, value):
        if not value:
            raise ValueError
        self.year = value

    def getVendor(self):
        return self.vendor

    def setVendor(self, value):
        if not value:
            raise ValueError
        self.vendor = value

    def getPower(self):
        return self.power

    def setPower(self, value):
        if not value:
            raise ValueError
        self.power = value

    def getColor(self):
        return self.color

    def setColor(self, value):
        if not value:
            raise ValueError
        self.color = value

    def getCost(self):
        return self.cost

    def setCost(self, value):
        if not value:
            raise ValueError
        self.cost = value

    def __eq__(self, other):
        return self.name == other.name and self.year == other.year\
            and self.vendor == other.vendor and self.power == other.power\
            and self.color == other.color\
            and self.cost == other.cost
    
    def __hash__(self):
        return hash((self.name.lower(), self.year, self.vendor.lower(), self.power, self.color.lower(), self.cost))
```
